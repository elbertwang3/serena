```python
import pandas as pd
```

```python
sd = pd.read_csv('public/data/servedirection.csv')
```

```python
sd['match_id'] = sd['match_id'].str.split("-")
```

```python
sd.head()
```

```python
sd['row'] = sd['row'].str.split(" ")
```

```python
sd.head()
```

```python
sd['playernumber'] = sd['row'].str[0]
```

```python
sd.head()
```

```python
for index, row in sd.iterrows():
    if row['playernumber'] == "1":
        sd.set_value(index, 'player', row['match_id'][4])
    else:
        sd.set_value(index, 'player', row['match_id'][5])
        
```

```python
sd.head()
```

```python
sd['serve'] = sd['row'].str[1]
```

```python
sd.head()
```

```python
sd['player'] = sd['player'].str.replace('_',' ')
```
